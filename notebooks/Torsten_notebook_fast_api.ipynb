```python
import requests
import pandas as pd
import datetime
```

```python


# API Endpoint URL
base_url = "http://127.0.0.1:8000/predict_snowfall?lat=47.26580883196723&long=11.84457426992035"

# Input Parameters for Prediction
latitude = 47.2658  # Example latitude
longitude = 11.84  # Example longitude

# Make the API Request
params = {"lat": latitude, "long": longitude}

try:
    # Sending the GET request to FastAPI
    response = requests.get(base_url, params=params)

    # Check if the request was successful
    if response.status_code == 200:
        result = response.json()
        print("Prediction Response:", result)
        print(f"Predicted Snowfall: {result['prediction']}")
    else:
        print(f"Error {response.status_code}: {response.text}")
except Exception as e:
    print("Error connecting to API:", e)
```
